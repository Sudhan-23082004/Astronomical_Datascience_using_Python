```python
# Import numpy
import numpy as np
```

```python
# Create an array called x of values between -5 to 5 with 50 points in between (equispaced)
x = np.linspace(-5, 5, 50)

# Create another array called y using the array x, which is the square of values in x array
y = x**2 # x to the power of 2
```

```python
# Display the arrays and check the type
print("x:-")
print(x)
print()
print("y:-")
print(y)
print()
print(f'Type of x:- {type(x)}')
print(f'Type of y:- {type(y)}')
```

```python
# Check the data type of the array x and y
print(f'Data type of x array:- {x.dtype}\nData type of y array:- {y.dtype}')
```

```python
# Import matplotlib library
import matplotlib.pyplot as plt
```

```python
# Create a scatter plot using above x and y and customize it using labels and title
plt.scatter(x, y, c='maroon')
plt.xlabel('X - Values', color='magenta', fontsize=11)
plt.ylabel('Y - Values', color='magenta', fontsize=11)
plt.title('$Y = x^2$ Visualization', color = 'royalblue', fontsize=15, weight='bold')
plt.show()


```

```python
# Sublot of same x and y (2 x 2), but with different colors, markers and linestyle

plt.figure(figsize=(10, 6))

plt.subplot(221)
plt.plot(x, y, 'r*-')
plt.subplot(222)
plt.plot(x, y, 'go--')
plt.subplot(223)
plt.plot(x, y, 'bx--')
plt.subplot(224)
plt.plot(x, y, 'ys-')

plt.show()


```

```python
# Create three different plots (x, x^2), (x, x^3), (X, x^4), add legend and save the output (customize it)
plt.plot(x, x**2, 'b.-', label='Quadratic')
plt.plot(x, x**3, 'g--', label='Cube')
plt.plot(x, x**4, 'm-', label='Quartic')
plt.legend(loc='upper center')
plt.savefig('my_first_plot.png')
plt.show()
```

```python
# Now create (x, x+9) to (x, x+12) four graphs and seperately add linestyle, linewidth, marker, and colors
plt.figure(figsize=(12, 6))
#plt.style.use('default')
plt.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
plt.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
plt.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
plt.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
plt.show()
```

```python
# Crop the important parts from the image using array slicing concept (can create a subplot if you want to show all in one figure)

plt.subplot(121)
plt.imshow(img[55:130, 50:150])
plt.axis('off')
plt.title('Pleiades', color='tab:blue', weight='bold')
plt.subplot(122)
plt.imshow(img[180:240, 200:280])
plt.axis('off')
plt.tight_layout()
plt.title('Moon', color='gray', weight='bold')
plt.savefig('PleiadesAndMoonExtracted.png')
plt.show()
```
