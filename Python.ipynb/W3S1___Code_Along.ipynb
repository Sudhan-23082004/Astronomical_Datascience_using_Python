```python
# Import numpy

import numpy as np
```

```python
# Create an array called x of values between -5 to 5 with 50 points in between (equispaced)
x =np.linspace(-5, 5, 50)

# Create another array called y using the array x, which is the square of values in x array
y = x**2
```

```python
# Display the arrays and check the type
print("The array x is: ")
print(x)
print("The array y is: ")
print(y)
print("\nThe type of x is: ",type(x))
print("The type of y is: ",type(y))

```

```python
# Import matplotlib library
import matplotlib.pyplot as plt

```

```python
# Create a scatter plot using above x and y and customize it using labels and titles
plt.scatter(x,y, color="red", marker="o")
plt.xlabel("X values", color="magenta")
plt.ylabel("Y values", color="magenta")
plt.title("Scatter plot of $y = x^2$", color="black", weight="bold",style="italic")
plt.show()

```

```python
# Subplot of same x and y (2 x 2), but with different colors, markers and linestyle
plt.figure(figsize=(11,9))
plt.subplot(2,2,1)
plt.plot(x,y, 'rs-')
plt.subplot(2,2,2)
plt.plot(x,y, 'bo-')
plt.subplot(2,2,3)
plt.plot(x,y, 'gp-')
plt.subplot(2,2,4)
plt.plot(x,y, 'y*-')
plt.show()

```

```python
# Create three different plots (x, x^2), (x, x^3), (X, x^4), add legend and save the output (customize it)

plt.plot(x, x**2, 'r.-', label="Quadratic")
plt.plot(x, x**3, 'm--', label="Cubic")
plt.plot(x, x**4, 'y-', label="Quartic")
plt.legend(loc="upper center")
plt.savefig("plot.png")
plt.show()
```

## Visualizing Image Data

```python
# Import imread from skimage.io

from skimage.io import imread
```

```python
image_path = 'C:\\Users\\sudha\\OneDrive\\Pictures\\Screenshots\\Screenshot 2023-12-08 221806.png'
img = imread(image_path)
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
image_path = 'C:\\Users\\sudha\\OneDrive\\Pictures\\Screenshots\\Screenshot 2023-12-08 221806.png'
img = imread(image_path, as_gray=True)
print(img.shape)
plt.imshow(img, cmap="gray")
plt.show()
```

```python
# Crop the important parts from the image using array slicing concept (can create a subplot if you want to show all in one figure)

plt.figure(figsize=(11,9))
plt.subplot(1,2,1)
plt.imshow(img[60:280, 50:542])
plt.axis("off")
plt.title("Current Weather", color="tab:blue", weight="bold")
plt.tight_layout()
plt.subplot(1,2,2)
plt.imshow(img[398:550, 815:957])

plt.title("Forecast", color="tab:red", weight="bold")  
plt.axis("off")
plt.show()
```
