```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage.io import imread
```

```python
# Loading the Jupiter Image
jupiter_img = cv2.imread("C:\\Spartificial\\Images\\gallery_338887_14764_2894.png")
```

```python
# Split will extract all the channels in the input image--> B, G, R
B, G, R = cv2.split(jupiter_img)
```

```python
# Check shape of B, G, R

print(f'Shapes:- {B.shape, G.shape, R.shape}')
```

```python
tuple_of_channels = (B, G, R)
titles = ('Blue Channel', 'Green Channel', 'Red Channel')
colors = ('Blue', 'Green', 'Red')

plt.figure(figsize=(13, 5))
plt.suptitle('Visualizing different channels of the Jupiter Image\n', weight='bold')

for i, channel in enumerate(tuple_of_channels):
  plt.subplot(1, 3, i+1)
  plt.imshow(channel, cmap='gray')
  plt.title(titles[i], color=colors[i])
  plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Check the data type
jupiter_img.dtype
```

```python
# Create an array of zeros of the same shape as the channels shape
zeros_arr = np.zeros(R.shape, dtype=np.uint8)
```

```python
def plot_merge_channels(first_channel, second_channel, third_channel,title):
  '''
  Params
  ------
  Each channels are a 2D Numpy Array

  Output
  ------
  Plot of 3 channel merged image
  '''
  plt.imshow(cv2.merge([first_channel, second_channel, third_channel]))
  plt.title(title)
  plt.tight_layout()
  plt.axis('off')
  plt.show()
```

```python
# Call this function to display different mergings of 3 channels
plot_merge_channels(R, zeros_arr, zeros_arr, title='Red Channel Given')
plot_merge_channels(zeros_arr, G, zeros_arr, title='Green Channel Given')
plot_merge_channels(zeros_arr, zeros_arr, B, title='Blue Channel Given')
plot_merge_channels(R, G, B, title='RGB Image')
```

```python
# Using skimage to plot subplot of R, G, B channels from the fruits and vegitables

def plot_rgb_channels(image_path):
  """
  Plotting the RGB channels of an image using subplots.

  Parameters:
  -----------
  image_path (str): The path to the image file.

  """
  # Read the image
  fruit_veg_arr = imread(image_path)

  # Define the titles for the subplots
  titles = ('Original Image', 'Red Channel', 'Green Channel', 'Blue Channel')

  # Create a figure with a size of 10x7 inches
  plt.figure(figsize=(10, 7))

  # Iterate over the channels and plot them
  for i in range(4): # 0, 1, 2, 3
    if i == 0:
      # Plot the original image
      plt.subplot(2, 2, i+1)
      plt.imshow(fruit_veg_arr)
      plt.title(f'{titles[i]}\n{fruit_veg_arr[:, :, i].shape}')
      plt.axis('off')
    else:
      # Plot the individual color channels
      plt.subplot(2, 2, i+1)
      plt.imshow(fruit_veg_arr[:, :, i-1], cmap='viridis')
      plt.title(f'{titles[i]}\n{fruit_veg_arr[:, :, i-1].shape}')
      plt.axis('off')

  # Adjust the layout of the subplots
  plt.tight_layout(pad=0, w_pad=1.0, h_pad=1.0)

  # Display the plot
  plt.show()

# Call the function
plot_rgb_channels('C:\\Spartificial\\Images\\367271.jpg')
```
